```python
import pickle
```

```python
with open('fNet-stdp-3000-epoch-loss-accuracy.pkl', 'rb') as f:
  o = pickle.load(f)
```

```python
o
```

```python
import matplotlib.pyplot as plt
```

```python
test_accuracies = o['test_accuracies']
```

```python
x = []
for i in range(len(test_accuracies)):
    x.append(i*10)
    test_accuracies[i] = float(test_accuracies[i])
```

```python
# plotting the points 
plt.plot(x, test_accuracies)
  
# naming the x axis
plt.xlabel('Epoches')
# naming the y axis
plt.ylabel('Accuracy')
  
# giving a title to my graph
plt.title('STDP accuracy by epoch')
  
# function to show the plot
plt.show()
```
