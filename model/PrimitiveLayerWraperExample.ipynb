```python
import torch
import torch.nn as nn
class PrimitiveLayerWraper(torch.nn.Module):
    def __init__(self, layer):
        super().__init__()
        self.layer = layer
        self.layer.requires_grad = False
    
    def forward(self, x):
        x = x.detach()
        x.requires_grad = False
        output = self.layer(x)
        output.detach()
        return output.detach()

```

```python
a = nn.Linear(10, 10)
a.requires_grad = True

b_a = nn.Linear(10, 10)
b_a.requires_grad = True

b = PrimitiveLayerWraper(b_a)

c = nn.Linear(10, 10)
c.requires_grad = True

model = nn.Sequential(a, b , c)

output = model(torch.ones(10))
output.backward(gradient=torch.ones(10))
print("Primitive layer gradients is", b_a.weight.grad)
```
