```python
import torch
import time
```

```python
totalTaskAmount = 10000

x = torch.rand((5000, 5000)).cuda()
m = torch.rand((5000, 5000)).cuda()
start = time.time()
for i in range(totalTaskAmount):
    a = torch.matmul(x, m)
end = time.time()
print('single thread time usage', end - start)
```

```python
from threading import Thread


class T:
    def __init__(self):
        self.x = torch.rand((5000, 5000)).cuda()
        self.m = torch.rand((5000, 5000)).cuda()
    def run(self):
        return torch.matmul(self.x, self.m) 
def update(t):
    for i in range(10000 // 2):
        t.run()
t = T()
start = time.time()
p1 = Thread(target=update, args=[t])
p1.start()
p2 = Thread(target=update, args=[t])
p2.start()
p1.join() 
p2.join()
end = time.time()
print('double thread time usage', end - start)

```
